```python
import pandas as pd
```

Load Dataset

```python
df=pd.read_csv('/content/laptop_price - dataset.csv')
df
```

Encoding the features

```python
from sklearn.preprocessing import LabelEncoder


df['Product'] = LabelEncoder().fit_transform(df['Product'])
df['Company'] = LabelEncoder().fit_transform(df['Company'])
df['CPU_Type'] = LabelEncoder().fit_transform(df['CPU_Type'])
df['RAM (GB)'] = LabelEncoder().fit_transform(df['RAM (GB)'])
df
```

```python
from sklearn.preprocessing import StandardScaler

x=df[['Product','Company','RAM (GB)','CPU_Type']]
scaler = StandardScaler()
x = scaler.fit_transform(x)

y=df['Price (Euro)']

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model=LinearRegression()
model.fit(x_train,y_train)
```

```python
x_test
```

```python
y_predict=model.predict(x_test)
y_predict
```

Evaluating the model accuracy

```python

r2 = r2_score(y_test, y_predict)
print(f"R^2 Score: {r2:.2f}")
```
